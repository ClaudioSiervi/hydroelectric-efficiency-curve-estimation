## Avaliação com diferentes métricas

* Quanto maior rsquared melhor é o ajuste, pois o modelo representa a a maior parte da variabilidade da amostra.
* Quanto maior rsauared_adj melhor é o ajuste, pois o modelo representa a a maior parte da variabilidade da amostra.  O ponto forte do rsauared_adj é que esta estatistica não aumenta de valor conforme aumenta o número de variáveis exploratórias no modelo.

* Quanto menor o valor de PRESS melhor é o ajuste do modelo avaliado.

* The F value in regression is the result of a test where the null hypothesis is that all of the regression coefficients are equal to zero. In other words, the model has no predictive capability. Basically, the f-test compares your model with zero predictor variables (the intercept only model), and decides whether your added coefficients improved the model. If you get a significant result, then whatever coefficients you included in your model improved the model’s fit. Read your p-value first. If the p-value is small (less than your alpha level), you can reject the null hypothesis. Only then should you consider the f-value. If you don’t reject the null, ignore the f-value.
https://www.statisticshowto.com/probability-and-statistics/f-statistic-value-test/
